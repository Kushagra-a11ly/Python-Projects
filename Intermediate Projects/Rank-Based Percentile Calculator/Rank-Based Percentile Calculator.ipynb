```python
import tkinter as tk
from tkinter import messagebox

def calculate_percentile():
    try:
        rank = int(rank_entry.get())
        total = int(total_entry.get())
        
        if rank <= 0 or rank > total:
            messagebox.showerror("Invalid Input", "Rank must be between 1 and Total Students.")
            return
        percentile = ((total - rank) / total) * 100
        result_label.config(text=f"Percentile: {percentile:.2f}")
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid integers for Rank and Total Students.")


root = tk.Tk()
root.title("Rank-Based Percentile Calculator")
root.geometry("400x250")
root.resizable(False, False)

title_label = tk.Label(root, text="Rank-Based Percentile Calculator", font=("Helvetica", 16, "bold"))
title_label.pack(pady=10)

rank_frame = tk.Frame(root)
rank_frame.pack(pady=5)
rank_label = tk.Label(rank_frame, text="Rank: ", font=("Helvetica", 12))
rank_label.pack(side=tk.LEFT)
rank_entry = tk.Entry(rank_frame, width=10, font=("Helvetica", 12))
rank_entry.pack(side=tk.LEFT)

total_frame = tk.Frame(root)
total_frame.pack(pady=5)
total_label = tk.Label(total_frame, text="Total Students: ", font=("Helvetica", 12))
total_label.pack(side=tk.LEFT)
total_entry = tk.Entry(total_frame, width=10, font=("Helvetica", 12))
total_entry.pack(side=tk.LEFT)

calc_button = tk.Button(root, text="Calculate Percentile", font=("Helvetica", 12, "bold"), bg="green", fg="white", command=calculate_percentile)
calc_button.pack(pady=15)

result_label = tk.Label(root, text="Percentile: ", font=("Helvetica", 14, "bold"))
result_label.pack(pady=10)

root.mainloop()
```
