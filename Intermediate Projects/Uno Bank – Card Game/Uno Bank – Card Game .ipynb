```python
import random
import time

# CARD SETUP
def build_single_player_deck():
    deck = [
        "+1", "+3", "+5", "+7",            
        "-2", "-4", "-6", "-8",            
        "x2", "/2", "+2D", "-2D"           
    ]
    return deck.copy()


def build_two_player_deck():
    deck = [
        "+1", "+3", "+5", "+7",
        "-2", "-4", "-6", "-8",
        "x2", "/2", "+2D", "-2D",
        "SWAP", "FORWARD", "BACKWARD", "BOMB"    # Yellow cards
    ]
    return deck.copy()

# SINGLE PLAYER GAME

def play_single_player():
    print("\n----- UNO BANK: SINGLE PLAYER MODE -----\n")

    deck = build_single_player_deck()
    balance = 10
    draws = 10

    while draws > 0:
        card = random.choice(deck)
        print(f"Draw {11 - draws}: You got ‚Üí {card}")

        if card.startswith("+") and card.endswith("D"):
            draws += int(card[:-1])
            print(f"Extra draws added! Total draws = {draws}")

        elif card.startswith("-") and card.endswith("D"):
            draws += int(card[:-1])
            print(f"Lost draws! Total draws = {draws}")

        elif card.startswith("+"):
            balance += int(card)
            print(f"Added money! Balance = ${balance}")

        elif card.startswith("-"):
            balance += int(card)  # negative value
            print(f"Lost money! Balance = ${balance}")

        elif card == "x2":
            balance *= 2
            print(f"Balance doubled! Balance = ${balance}")

        elif card == "/2":
            balance //= 2
            print(f"Balance halved! Balance = ${balance}")

        draws -= 1
        time.sleep(0.5)

    print("\nFinal Balance:", balance)
    print("Game Over!")

# TWO PLAYER MODE

def apply_yellow_effect(card, p1_balance, p2_balance):
    if card == "SWAP":
        print(" SWAP CARD! Players swap balances.")
        return p2_balance, p1_balance

    elif card == "FORWARD":
        print("FORWARD CARD! You give all your money to opponent.")
        return 0, p1_balance + p2_balance

    elif card == "BACKWARD":
        print("BACKWARD CARD! You steal all money from opponent.")
        return p1_balance + p2_balance, 0

    elif card == "BOMB":
        print("BOMB CARD! Your balance resets to $0.")
        return 0, p2_balance

    return p1_balance, p2_balance


def play_two_player():
    print("\n----- UNO BANK: TWO PLAYER MODE -----\n")

    deck = build_two_player_deck()
    p1_balance = 10
    p2_balance = 10
    p1_draws = 10
    p2_draws = 10

    turn = 1  

    while p1_draws > 0 and p2_draws > 0:
        print(f"\n--- Player {turn}'s Turn ---")

        card = random.choice(deck)
        print(f"Player {turn} drew ‚Üí {card}")

        if turn == 1:
            if card in ["SWAP", "FORWARD", "BACKWARD", "BOMB"]:
                p1_balance, p2_balance = apply_yellow_effect(card, p1_balance, p2_balance)

            elif card.startswith("+") and card.endswith("D"):
                p1_draws += int(card[:-1])
                print(f"Player 1 gains draws ‚Üí {p1_draws}")

            elif card.startswith("-") and card.endswith("D"):
                p1_draws += int(card[:-1])
                print(f"Player 1 loses draws ‚Üí {p1_draws}")

            elif card.startswith("+"):
                p1_balance += int(card)
            elif card.startswith("-"):
                p1_balance += int(card)
            elif card == "x2":
                p1_balance *= 2
            elif card == "/2":
                p1_balance //= 2

            p1_draws -= 1

        else:
            if card in ["SWAP", "FORWARD", "BACKWARD", "BOMB"]:
                p2_balance, p1_balance = apply_yellow_effect(card, p2_balance, p1_balance)

            elif card.startswith("+") and card.endswith("D"):
                p2_draws += int(card[:-1])
                print(f"Player 2 gains draws ‚Üí {p2_draws}")

            elif card.startswith("-") and card.endswith("D"):
                p2_draws += int(card[:-1])
                print(f"Player 2 loses draws ‚Üí {p2_draws}")

            elif card.startswith("+"):
                p2_balance += int(card)
            elif card.startswith("-"):
                p2_balance += int(card)
            elif card == "x2":
                p2_balance *= 2
            elif card == "/2":
                p2_balance //= 2

            p2_draws -= 1

        print(f"Balances ‚Üí P1: ${p1_balance} | P2: ${p2_balance}")

        turn = 2 if turn == 1 else 1
        time.sleep(0.5)

    print("\n----- GAME OVER -----")
    print(f"Final Balances ‚Üí P1: ${p1_balance} | P2: ${p2_balance}")

    if p1_balance > p2_balance:
        print("üéâ Player 1 Wins!")
    elif p2_balance > p1_balance:
        print("üéâ Player 2 Wins!")
    else:
        print("ü§ù It's a Draw!")

# MAIN MENU

def main():
    print("\n====== UNO BANK GAME ======\n")
    print("1. Single Player")
    print("2. Two Player")
    mode = input("\nChoose mode (1/2): ")

    if mode == "1":
        play_single_player()
    elif mode == "2":
        play_two_player()
    else:
        print("Invalid choice.")

main()
```
