```python
class PasswordChecker:

    def __init__(self, password):
        self.password = password

    def check_strength(self):
        strength = 0

        if len(self.password) >= 8:
            strength += 1
        if any(c.islower() for c in self.password):
            strength += 1
        if any(c.isupper() for c in self.password):
            strength += 1
        if any(c.isdigit() for c in self.password):
            strength += 1
        if any(c in "!@#$%^&*()-_=+{}[]|;:'\",.<>?/" for c in self.password):
            strength += 1

        # Strength Levels
        if strength <= 2:
            return "Weak Password"
        elif strength == 3:
            return "Medium Password"
        else:
            return "Strong Password"

user_password = input("Enter your password: ")

checker = PasswordChecker(user_password)
result = checker.check_strength()

print("Strength:", result)
```
