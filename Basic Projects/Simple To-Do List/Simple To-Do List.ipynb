```python
class ToDoList:

    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)
        print("Task added!")

    def show_tasks(self):
        print("\nYour Tasks:")
        for i, t in enumerate(self.tasks, 1):
            print(i, "-", t)

    def run(self):
        while True:
            print("\n1. Add Task\n2. Show Tasks\n3. Exit")
            choice = input("Enter choice: ")

            if choice == "1":
                task = input("Write task: ")
                self.add_task(task)
            elif choice == "2":
                self.show_tasks()
            else:
                break

todo = ToDoList()
todo.run()
```
