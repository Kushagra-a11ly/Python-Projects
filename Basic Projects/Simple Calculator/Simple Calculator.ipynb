```python
class Calculator:
    def get_input(self):
        self.num1 = float(input("Enter the first number: "))
        self.operator = input("Enter the operator (+, -, *, /): ")
        self.num2 = float(input("Enter the second number: "))

        # New line you requested
        print(f"Number 1 is: {self.num1} and Number 2 is: {self.num2}")

    def calculate(self):
        if self.operator == "+":
            result = self.num1 + self.num2
            print(f"Addition of {self.num1} and {self.num2} is: {result}")

        elif self.operator == "-":
            result = self.num1 - self.num2
            print(f"Subtraction of {self.num1} and {self.num2} is: {result}")

        elif self.operator == "*":
            result = self.num1 * self.num2
            print(f"Multiplication of {self.num1} and {self.num2} is: {result}")

        elif self.operator == "/":
            if self.num2 == 0:
                print("Error: Division by zero is not allowed.")
            else:
                result = self.num1 / self.num2
                print(f"Division of {self.num1} by {self.num2} is: {result}")

        else:
            print("Invalid operator entered.")

calc = Calculator()
calc.get_input()
calc.calculate()
```
