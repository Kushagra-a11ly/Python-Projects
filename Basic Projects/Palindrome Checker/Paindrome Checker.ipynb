```python
def is_palindrome(text):
    cleaned_text = text.replace(" ", "").lower()
    return cleaned_text == cleaned_text[::-1]

def main():
    print("üîÑ Welcome to the Palindrome Checker!")
    
    while True:
        user_input = input("Enter a word or phrase: ")
        if is_palindrome(user_input):
            print(f"‚úÖ'{user_input}' is a palindrome!\n")
        else:
            print(f"‚ùå '{user_input}' is NOT a palindrome.\n")
        
        again = input("Do you want to check another? (yes/no): ").lower()
        if again != "yes":
            print("Thank you for using the Palindrome Checker. Goodbye!")
            break

if __name__ == "__main__":
    main()
```
